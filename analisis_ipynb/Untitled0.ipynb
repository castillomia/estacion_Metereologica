```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
df = pd.read_csv('dataset_limpio.csv')

# Configurar el estilo de los gráficos
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

print("¡Dataset cargado con éxito!")
df.head() # Muestra las primeras filas
```

```python
# 📊 Comparación 1:
sns.countplot(data=df, y='Criterio de Seguridad Climática', palette='viridis')
plt.title('Distribución de Factores Críticos de Decisión')
plt.show()

```

Este gráfico nos dice qué sensor de nuestra estación es el más "esperado" por la comunidad.
**Comentario:** "Si la barra de 'Precipitaciones' es la más alta, nuestra estación debe ser infalible detectando lluvia, porque es lo que define si se hace o no una actividad".


```python
# 📊 Comparación 2:
sns.countplot(data=df, x='Solución de Valor', palette='magma')
plt.xticks(rotation=45)
plt.title('Principales "Dolores" que la Estación debe resolver')
plt.show()
```

Aquí vemos la utilidad real (Product Discovery) de nuestro invento.
**Comentario:** "Si 'Salud y Bienestar' lidera, la estación no solo da el clima, sino que protege a los alumnos de alergias o golpes de calor".


```python
# 📊 Comparación 3:
sns.barplot(data=df, x='Edad', y='Prevencion_Ambiental', estimator=lambda x: (x == 'Sí').mean() * 100)
plt.ylabel('Porcentaje que eligió Cuidado Ambiental (%)')
plt.title('Conciencia Ambiental por Rango Etario')
plt.show()
```

Preocupación Ambiental según la Edad
¿A qué edad son más conscientes del impacto de la huella de carbono?
**Comentario:** "Este gráfico nos muestra si los más grandes o los más chicos están más alineados con el trabajo de los profes de Biología".


```python
# 📊 Comparación 4:
paralisis_counts = df['Prevencion_Info'].value_counts()
plt.pie(paralisis_counts, labels=paralisis_counts.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
plt.title('¿Desean evitar el exceso de información técnica?')
plt.show()
```

Interés en Evitar la "Parálisis por Información"
¿Realmente necesitan una interfaz simple o les gustan los números complejos?
